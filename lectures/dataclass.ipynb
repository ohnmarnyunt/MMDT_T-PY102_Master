```python
class student():
    def __init__(self, new_name: str, new_id: str):
        self.name = new_name
        self.id = new_id

stu1 = student("myo", "2026001")
stu2 = student("myo", "2026001")

print(stu1==stu2)

print(id(stu1))
print(id(stu2))

print(stu1)
```

```python
from dataclasses import dataclass

@dataclass
class student:
    name:str
    id: str

stu1 = student("Myo", "2026001")
stu2 = student("Myo", "2026001")
print(stu1)

print(stu1==stu2)
```

```python
from dataclasses import dataclass

@dataclass
class student:
    name:str = "MMDT"
    id: str = "2026000"

s = student()
print(s)
```

```python
from dataclasses import dataclass, field

@dataclass
class student:
    name: str = "MMDT"
    id: str = "2026000"
    pw: str= field(default = "@@@mmdt", repr = False)

stu = student()
print(stu)
```

field() for advanced defaults and options
4.1 Mutable defaults (lists/dicts/sets)

```python
from dataclasses import dataclass

@dataclass
class student:
    name:str
    id: str = "2026000"
```

```python
@dataclass
class Student:
    name: str
    id: str = "2026000"
    subjects: list = field(default_factory=list)

s1 = Student("A")
s2 = Student("B")

s1.subjects.append("math")
print(s1)
print(s2)
```
